# Demo de inferencia sencilla
Este notebook simula el flujo mínimo: se entrena un modelo liviano de scikit-learn y se guarda en disco, luego se carga ese modelo para hacer una inferencia rápida.

### Instrucciones
1. Asegúrate de tener el entorno activado (`python -m venv .venv`).
2. Instala las dependencias: `pip install -r requirements.txt`.
3. Ejecuta cada celda (Shift+Enter) y observa cómo se genera el archivo `examples/models/rf-demo.joblib`.
4. El bloque final muestra cómo cargar ese modelo y realizar una predicción de ejemplo.

```python
from pathlib import Path
from joblib import dump
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier

Path('examples/models').mkdir(parents=True, exist_ok=True)
iris = load_iris()
clf = RandomForestClassifier(n_estimators=10, random_state=42)
clf.fit(iris.data, iris.target)
dump(clf, 'examples/models/rf-demo.joblib')
print('Modelo entrenado y guardado en examples/models/rf-demo.joblib')
```

```python
from joblib import load
from sklearn.datasets import load_iris
import numpy as np

model = load('examples/models/rf-demo.joblib')
sample = load_iris().data[:1]
prediction = model.predict(sample)
print('Predicción para el primer dato del Iris:', prediction[0])
print('Etiquetas del dataset:', load_iris().target_names)
```
